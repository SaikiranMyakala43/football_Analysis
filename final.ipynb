```python
#lets find some critical information to find missing, unique and count values
#Return a Pandas dataframe describing the contents of a source dataframe including missing values
def show_missing(football):
    
    variables = []
    dtypes = []
    count = []
    unique = []
    missing = []
    pc_missing = []
    
    for item in football.columns:
        variables.append(item)
        dtypes.append(football[item].dtype)
        count.append(len(football[item]))
        unique.append(len(football[item].unique()))
        missing.append(football[item].isna().sum())
        pc_missing.append(round((football[item].isna().sum() / len(football[item])) * 100, 2))

    output = pd.DataFrame({
        'variable': variables, 
        'dtype': dtypes,
        'count': count,
        'unique': unique,
        'missing': missing, 
        'pc_missing': pc_missing
    })    
    return output
print(show_missing(football))
#we can see lot of variables has the missing values so we lets stic to the questionaires we have given so we can have more focus
```

```python
#lest remove all the columns which does not make any impact on our analysis
columns_to_delete = ['Unnamed: 0','Photo','Flag','Club Logo','LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM', 'RCM','RM','LWB','LDM', 'CDM','RDM','RWB','LB','LCB','CB','RCB','RB']
football = football.drop(columns=columns_to_delete)
```

```python
football['Position'].fillna(football['Position'].mode()[0], inplace=True)
```

```python
football.head(5)
```

```python
football['Value'] = football['Value'].str.replace('€', '')
football['Value'] = football['Value'].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
```

```python
football['Wage'] = football['Wage'].str.replace('€', '')
football['Wage'] = football['Wage'].replace({"K":"*1e3", "M":"*1e6"}, regex=True).map(pd.eval).astype(int)
```

```python
football.head()
```

```python
# Preprocess the 'Height' 
def preprocess_height(height):
    if pd.isna(height) or height == 'nan':
        return np.nan
    else:
        feet, inches = map(int, height.split("'"))
        total_inches = feet * 12 + inches
        return total_inches / 12.0

# Apply the preprocessing function to the 'Height' column
football['Height'] =football['Height'].apply(preprocess_height)

# Replace "nan" values with the mode
height_mode = football['Height'].mode()[0]
football['Height'].fillna(height_mode, inplace=True)

# Convert the 'Height' column to numerical format
football['Height'] = pd.to_numeric(football['Height'])
football.head()
```

```python
#lets preprocess the weight
def preprocess_weight(weight):
    if pd.isna(weight) or weight.lower() == 'nan':
        return np.nan
    else:
        return float(weight.lower().replace('lbs', ''))
    
# Apply the preprocessing function to the 'Weight' column
football['Weight'] = football['Weight'].apply(preprocess_weight)

# Replace "nan" values with the mode
weight_mode = football['Weight'].mode()[0]
football['Weight'].fillna(weight_mode, inplace=True)

# Convert the 'Weight' column to numerical format
football['Weight'] = pd.to_numeric(football['Weight'])
football.head()
```

```python
# Univariate analysis for 'Height', i am using box plot beacuse it is easy to identify the outliers
# as points which falls significantly outside the whiskers of the box plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=football['Height'])
plt.title('Height - Box Plot')

# Univariate analysis for 'Weight'
plt.subplot(1, 2, 2)
sns.boxplot(x=football['Weight'])
plt.title('Weight - Box Plot')

plt.show()
#we can see there are lot of outliers that in both weight and height weioghts
```

```python
#another way of soing Univariate analysis is by IQR and this is most statistical way of approach 
# Calculate IQR for 'Height'
Q1_height = football['Height'].quantile(0.25)
Q3_height = football['Height'].quantile(0.75)
IQR_height = Q3_height - Q1_height

# Identify potential outliers for 'Height'
height_outliers = football[(football['Height'] < Q1_height - 1.5 * IQR_height) | (football['Height'] > Q3_height + 1.5 * IQR_height)]
print("\nPotential outliers for 'Height':")
print(height_outliers)

# Calculate IQR for 'Weight'
Q1_weight = football['Weight'].quantile(0.25)
Q3_weight = football['Weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight

# Identify potential outliers for 'Weight'
weight_outliers = football[(football['Weight'] < Q1_weight - 1.5 * IQR_weight) | (football['Weight'] > Q3_weight + 1.5 * IQR_weight)]
print("\nPotential outliers for 'Weight':")
print(weight_outliers)
```

```python
#Bivariate analysis
# Bivariate analysis for 'Height' and 'Weight'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Height', y='Weight', data=football)
plt.title('Bivariate Analysis: Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')

plt.show()
```

```python
from scipy.spatial.distance import mahalanobis

# Combine 'Height' and 'Weight' into a 2D array
data_array = football[['Height', 'Weight']].values

# Calculate the mean and covariance matrix
mean_values = np.mean(data_array, axis=0)
covariance_matrix = np.cov(data_array, rowvar=False)

# Calculate Mahalanobis distance for each data point
mahalanobis_distances = []
for row in data_array:
    distance = mahalanobis(row, mean_values, np.linalg.inv(covariance_matrix))
    mahalanobis_distances.append(distance)

# Convert the distances to a numpy array
mahalanobis_distances = np.array(mahalanobis_distances)

# Set a threshold for identifying outliers (adjust as needed)
threshold = 3.0

# Identify outliers based on Mahalanobis distance
outliers_indices = np.where(mahalanobis_distances > threshold)
outliers = football.iloc[outliers_indices]

print("\nOutliers based on Mahalanobis distance:")
print(outliers)
```
